# Typically, the x-axis represents the independent variable (e.g., 'Rooms' or 'Car'), and the y-axis represents the dependent variable (e.g., 'Price').

```python
random_seed = 42 # It provides reproducibility,Sets the random seed to 42. This ensures that if you run the code again, you'll get the same random subsample
subsampled_data = data.sample(frac=0.02, random_state=random_seed)
plt.figure(figsize=(20, 10))
sizes = [300]

# Add jitter to the x-axis positions
jittered_rooms = subsampled_data['Rooms'] + np.random.normal(0, 0.2, len(subsampled_data))
jittered_car = subsampled_data['Distance'] + np.random.normal(0, 0.2, len(subsampled_data))
plt.xticks(range(50))
plt.scatter(jittered_rooms, subsampled_data['Price'], label='Price vs Rooms', color='blue', s=sizes, marker='*')
plt.scatter(jittered_car, subsampled_data['Price'], label='Price vs Distance', color='orange', alpha = 0.5) 
plt.colorbar()
plt.title('Housing Price vs Rooms and Distance (Subsampled 10% with Seed)')
plt.xlabel('Rooms and Distance')
plt.ylabel('Price') 
plt.legend()
plt.show()
```

# Linear Regression

```python
# Importing necessary libraries 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Suburb'] = le.fit_transform(data['Suburb'])
data['Type'] = le.fit_transform(data['Type'])
data['Method'] = le.fit_transform(data['Method'])
data['SellerG'] = le.fit_transform(data['SellerG'])
data['CouncilArea'] = le.fit_transform(data['CouncilArea'])
data['Regionname'] = le.fit_transform(data['Regionname'])
data["Address"] = le.fit_transform(data['Address'])
data["Date"] = le.fit_transform(data['Date'])
```

```python
train = data.drop(['Price'], axis=1)
test = data['Price']
```

```python
x_train, x_test, y_train, y_test = train_test_split( train, test, test_size=0.2, random_state=7)
```

```python
regr = LinearRegression()
```

```python
regr.fit(x_train, y_train)
```

```python
pred = regr.predict(x_test)
```

```python
pred
```

```python
regr.score(x_test, y_test)
```

```python
plt.scatter(y_test, pred)
plt.title('Actual Price vs Predicted Price')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.show()
```
